```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/prieto-daniel/Data-Science-I/refs/heads/main/Datos/TH.csv?token=GHSAT0AAAAAADDOZ5WZXBR533A2WTBXS6AU2A3OUQA'
df = pd.read_csv(url, delimiter=';')
df.head()
```

```python
provincias:list[str] = df['PROVINCIA'].unique()
bp:list[int] = []

for provincia in provincias:
    #datos_provincia:list[int] = pd.to_numeric(df[df['PROVINCIA'] == provincia]['BP'], errors='coerce')
    datos_provincia:list[int] = df[df['PROVINCIA'] == provincia]['BP'].astype(int)
    #print(datos_provincia.head())
    bp.append(sum(datos_provincia))

data = {
    "PROVINCIA" : provincias,
    "BP" : bp
}

dataset = pd.DataFrame(data)

dataset.head()
```

```python
# Visualizar los datos
plt.figure(figsize = (16, 9))
sns.set_style("darkgrid")
#ax = sns.barplot(x = "PROVINCIA", y = "BP", data = df, estimator='sum')
#ax = sns.barplot(x = "PROVINCIA", y = "BP", data = dataset)
ax = sns.barplot(x = "PROVINCIA", y = "BP", data = dataset)
plt.title('Movimientos por Provincia', fontsize = 16)
plt.xlabel('PROVINCIAS', fontsize = 10)
plt.ylabel('MOVIMIENTOS', fontsize = 10)
plt.xticks(rotation = 45, ha = 'right')
plt.tight_layout()
ax.bar_label(ax.containers[0], fmt='{:,.0f}', fontsize=8)

# Añadir etiquetas de valor en cada barra
#for i, v in enumerate(dataset['BP']):
#    ax.text(i, v + 0.5, str(v), ha='center', fontsize=8)

plt.show()
```
